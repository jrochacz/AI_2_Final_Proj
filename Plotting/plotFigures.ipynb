```python
hist_data = loadObject('Results/Experiment1_pkl')
```

```python
actFuncList_str = ['relu', 'gelu', 'squarplus', 'softplus','erfrelu', 'erfplus', 'erfplus2'] 
 
trials = 5

epochs = 25 
```

```python
for ii in range(len(actFuncList_str)):
    Score_matrix = np.zeros((trials, epochs))
    for jj in range(trials): 
        accuracies = [x['val_acc'] for x in hist_data[jj][ii]]
        
        score = -np.log(1-np.array(accuracies))
        
        Score_matrix[jj, :] = score 
        
    plt.plot(np.mean(Score_matrix, axis=0), '-', label=actFuncList_str[ii])
    
plt.xlabel('epoch')
plt.ylabel('validation score')
plt.legend()
plt.grid()
plt.title(f'Validation score for activation functions with {trials} trials');
```

```python
for ii in range(len(actFuncList_str)):
    loss = [x['val_loss'] for x in hist_data[0][ii]]
    plt.plot(loss, '-', label=actFuncList_str[ii])
    
plt.xlabel('epoch')
plt.ylabel('validation accuracy')
plt.legend()
plt.grid()
plt.title('Validation accuracy for activation functions');
```
